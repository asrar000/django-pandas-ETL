# Raw Data Viewer

This notebook uses PySpark to view the raw data files from the ETL pipeline.

Files viewed:
- `data/raw/raw_carts.json`
- `data/raw/raw_users.json`
- `data/raw/synthesised_carts.json`
- `data/raw/synthesised_users.json`

```python
from services.spark_service import build_spark_session

spark = build_spark_session()
spark
```

```python
# View raw_carts.json
raw_carts_df = spark.read.json('data/raw/raw_carts.json')
print('Schema for raw_carts.json:')
raw_carts_df.printSchema()
print('\nSample data:')
raw_carts_df.show(5, truncate=False)
```

```python
# View raw_users.json
raw_users_df = spark.read.json('data/raw/raw_users.json')
print('Schema for raw_users.json:')
raw_users_df.printSchema()
print('\nSample data:')
raw_users_df.show(5, truncate=False)
```

```python
# View synthesised_carts.json
synthesised_carts_df = spark.read.json('data/raw/synthesised_carts.json')
print('Schema for synthesised_carts.json:')
synthesised_carts_df.printSchema()
print('\nSample data:')
synthesised_carts_df.show(5, truncate=False)
```
