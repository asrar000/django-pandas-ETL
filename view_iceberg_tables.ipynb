# Iceberg Tables Viewer

This notebook uses PySpark to view the Iceberg analytics tables after ETL execution.

Tables viewed:
- `local.analytics.customer_analytics`
- `local.analytics.order_analytics`

```python
from services.spark_service import build_spark_session

spark = build_spark_session()
spark
```

```python
# View customer_analytics table
customer_df = spark.sql("SELECT * FROM local.analytics.customer_analytics")
print('Schema for customer_analytics:')
customer_df.printSchema()
print('\nSample data:')
customer_df.show(5, truncate=False)
print(f'\nTotal rows: {customer_df.count()}')
```

```python
# View order_analytics table
order_df = spark.sql("SELECT * FROM local.analytics.order_analytics")
print('Schema for order_analytics:')
order_df.printSchema()
print('\nSample data:')
order_df.show(5, truncate=False)
print(f'\nTotal rows: {order_df.count()}')
```

```python
spark.stop()
```
