# Pandas PostgreSQL Viewer

This notebook uses Pandas to view the PostgreSQL analytics tables after ETL execution.

Tables viewed:
- `customer_analytics`
- `order_analytics`

```python
import pandas as pd
import psycopg2
from config.loader import get

def get_db_connection():
    return psycopg2.connect(
        host=get('database.host'),
        port=get('database.port'),
        dbname=get('database.name'),
        user=get('database.user'),
        password=get('database.password')
    )

def query_to_dataframe(query):
    conn = get_db_connection()
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df
```

```python
# SQL Queries
customer_query = """
SELECT * FROM customer_analytics LIMIT 10;
"""

order_query = """
SELECT * FROM order_analytics LIMIT 10;
"""

customer_count_query = """
SELECT COUNT(*) as total_customers FROM customer_analytics;
"""

order_count_query = """
SELECT COUNT(*) as total_orders FROM order_analytics;
"""
```

```python
# Execute queries
customer_df = query_to_dataframe(customer_query)
order_df = query_to_dataframe(order_query)
customer_count = query_to_dataframe(customer_count_query)
order_count = query_to_dataframe(order_count_query)
```

```python
# Display results
print("Customer Analytics Sample:")
display(customer_df)
print(f"\nTotal Customers: {customer_count.iloc[0,0]}")

print("\nOrder Analytics Sample:")
display(order_df)
print(f"\nTotal Orders: {order_count.iloc[0,0]}")
```
